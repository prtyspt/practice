```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
X, y = mglearn.datasets.make_wave(n_samples=40)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
reg = KNeighborsRegressor(n_neighbors=3)
```

```python
reg.fit(X_train, y_train)
```

```python
reg.predict(X_test)
```

```python
reg.score(X_test, y_test)
```

```python
fig, axes = plt.subplots(1, 3, figsize = (15, 4))
```

```python
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
```

```python
plt.suptitle("nearest neighbor regression")
```

```python
for n_neighbors, ax in zip([1, 3, 9], axes):
    reg = KNeighborsRegressor(n_neighbors=n_neighbors).fit(X_train, y_train)
    ax.plot(X_train, y_train, 'o')
    ax.plot(X_train, -3*np.ones(len(X_train)), 'o')
    ax.plot(line, reg.predict(line))
    ax.set_title("%d neigbors" %n_neighbors)
    
    
```

```python
fig
```

Two main parameters for K-nearest regression - No. of neigbors and Distance calculation method

Strength: easy to understand, reasonable performance, good baseline method, building it is fast, but if training set is high prediction can be slow.

Not good for datasets with many features, in particular sparse datasets.
